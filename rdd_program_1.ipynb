```python
import pyspark
print(pyspark.__version__)
from pyspark.sql import SparkSession
```

```python

spark = SparkSession \
    .builder \
        .appName("Python Spark create RDD Example") \
            .config("spark.some.config.option", "some-value") \
                .getOrCreate()

data = [(1,2,3,'a b c'),(4,5,6,'d e f'),(7,8,9,'g h i')]
rdd = spark.sparkContext.parallelize(data)
df = rdd.toDF(['col1', 'col2', 'col3', 'col4'])
            
```

```python
df.show()
```

# 2. By Using createDataFrame() function

```python
spark3 = SparkSession \
    .builder \
        .appName("Python Spark create RDD example.3") \
            .config("spark.some.config.option", "some-value") \
                .getOrCreate()

Employee = spark3.createDataFrame([
    ('1','Joe','70000','1'),
    ('2','Henry','80000','2'),
    ('3', 'Sam', '60000', '2'),
    ('4', 'Max', '90000', '1')
],['Id', 'Name', 'Sallary', 'DepertmentId'])

Employee.collect();
Employee.show();

```
